```python
import numpy as np
import cmath
import random
import matplotlib.pyplot as plt
def first(x,Bin,g):
    y=[]
    z=[]
    r=[]
    for i in x:
        a = complex(0.1,2*i)
        y.append(a)
        b = complex(1,2*i)
        z.append(b)
        ratio = (-1 + ((2*a)/(4*g*g + a*b)))
        r.append(ratio.real)
    ra=np.array(r)
    #fluc = sqrt(2)*10^(-4)*Bin*bin/KbT
    #fluc = 3.44 *(10^(16))*Bin
    #fluc = random.uniform(,)
     # + fluc/Bin 
    return ra
def second(x,Bin,g):
    y=[]
    z=[]
    r=[]
    fluc = random.uniform(-0.05,0.05)
    for i in x:
        a = complex(0.1,2*i)
        y.append(a)
        b = complex(1,2*i)
        z.append(b)
        ratio = (-1 + ((2*a)/(4*g*g + a*b))) + fluc
        r.append(ratio.real)
    ra=np.array(r)
    #fluc = sqrt(2)*10^(-4)*Bin*bin/KbT
    #
    #fluc = random.uniform(,)
     # + fluc/Bin 
    return ra

x = np.arange(start = -2,stop = 2,step = 0.01)
#Bin can't be zero ...but i assumed it to be zero because I don't know values of Bin..

plt.plot(x,second(x,0,0.225),label = 'spectra at g = 0.225', c = '#DC143C')
plt.plot(x,second(x,0,0.215),label = 'spectra at g = 0.215',c = '#00EEEE')
plt.plot(x,second(x,0,0.235),label = 'spectra at g = 0.235',c = '#76EE00')
plt.plot(x,second(x,0,0.245),label = 'spectra at g = 0.245',c = '#458B00')
plt.legend()
leg = plt.legend(loc='upper left')

plt.ylabel("Re(A_out/B_in)")
plt.xlabel("w-w'(k)")
plt.show()
```
